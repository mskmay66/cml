# Vizualizations 

```python
from cml import CoupledMapLattice, Visualization
import numpy as np
import seaborn as sns
```

```python
_map = CoupledMapLattice(10, 4, .75)
```

```python
states = _map.simulate(30)
```

```python
s = []
for state in states:
    s.append(state)
```

```python
viz = Visualization(_map)
```

```python
_map.history[0].shape
```

```python
viz.animate()
```

```python
nueron = 6

n_states = np.array(s)[:, nueron, nueron]
```

```python
n_states.shape
```

```python
n_states.shape
```

```python
s = np.array(s)
```

```python
corr = []
for i in range(len(s)):
    corr_start = n_states[i]
    corr_base = np.full((10, 10), corr_start)
    r = np.corrcoef(corr_base, s[i], rowvar=True)
    corr.append(r)
```

```python
len(corr[0])
```

```python
sns.heatmap(corr[2], cmap="YlGnBu", cbar=False)
```

```python
import matplotlib.pyplot as plt
from matplotlib import animation

fig = plt.figure()
data = corr[0]

def init():
      sns.heatmap(corr[0],cmap="YlGnBu", square=True, cbar=False)

def animate(i):
    data = corr[i]
    sns.heatmap(data, cmap="YlGnBu", square=True, cbar=False)

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=30, repeat = False)
```
